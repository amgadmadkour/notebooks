# Upload Tiny Shakespeare

## Load Modules

```python
from datasets import DatasetDict, Dataset
from pathlib import Path
```

## Load Data

```python
content = Path("../data/tiny_shakespeare/tiny-shakespeare.txt").read_text()
```

## Convert to HF

```python
dataset = DatasetDict({
    "train": Dataset.from_dict({"text": [content]}),
})
```

## Upload to HF Hub

```python
# Push to HF Hub
dataset.push_to_hub(
    "amgadmadkour/tiny_shakespeare"
)
print("Uploaded successfully!")
```
