# Upload Titanic

## Load Modules

```python
from datasets import DatasetDict, Dataset
import pandas as pd
```

## Load Data

```python
train_df = pd.read_csv("../data/titanic/train.csv")
test_df = pd.read_csv("../data/titanic/test.csv")
test_df["Survived"] = pd.Series([pd.NA] * len(test_df), dtype="Int64")
```

## Convert to HF

```python
dataset = DatasetDict({
    "train": Dataset.from_pandas(train_df),
    "test": Dataset.from_pandas(test_df),
})
```

## Upload to HF Hub

```python
# Push to HF Hub
dataset.push_to_hub(
    "amgadmadkour/titanic"
)
print("Uploaded successfully!")
```
